```python
def lookup_tweets(tweet_IDs, api):
    full_tweets = {}
    for tweet_ID in tweet_IDs:
        try:
            tweet = api.get_status(id=tweet_ID, tweet_mode= 'extended')
            full_tweets[tweet_ID] = tweet._json
        except tweepy.TweepError:
            full_tweets[tweet_ID]='ID Not Found'
            continue
    return full_tweets

consumer_key = 'FS4Fg3g0tXKjqcJqUGIQzF8rX'
consumer_secret = '6CI0vggQtsNzQSfaTA9rAqgraN1TZrPaPLt6KIgUXOIdYxCA0z'
access_token = '40641424-RpSYBgCn0H2i1u02UwQzRKho2afmklte05lgYNaM0'
access_token_secret = 'uV65tpOJ5M3Lfz8LeAfTZCDcKzJ8eteq6PLnxWqlFzupW'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)

list_tweet = lookup_tweets(list(twit_arc['tweet_id']), api)
```

```python
def get_names(sent):    
    try:    
        names = [word
                 for sent in sent_tokenize(sent)
                 for word in word_tokenize(sent)[1:]
                 if (word[0].isupper() and len(word)>1)]
        return names[0]
    except:
        return np.nan

twit_arc_copy['name'] = twit_arc_copy['text'].apply(lambda x: get_names(x))
```

Transform `doggo`, `floofer`, `pupper` & `puppo` columns into one column `dog_stage`

###### Code 

```python
def pupper(row):
    if row['doggo']=='doggo':
        return 'doggo'
    elif row['floofer']=='floofer':
        return 'floofer'
    elif row['pupper']=='pupper':
        return 'pupper'
    elif row['puppo']=='puppo':
        return 'puppo'
    else:
        return np.nan

twit_arc_copy['dog_stage'] = twit_arc_copy.apply(lambda x: pupper(x), axis=1)
    
```

We see that there are no instances of `p2_conf` being greater than `p1_conf` or `p3_conf` being greater than `p2_conf`. Safe to assume that to extract dog name, we need the highest confidence level with `p#_dog` as True

```python
def confidence_def(row):
    if row['p1_dog'] == True:
        return row['p1']
    else:
        if row['p2_dog'] == True:
            return row['p2']
        else:
            return row['p3']

image_predic_copy['dog_type'] = image_predic_copy.apply(lambda x: confidence_def(x), axis=1)
```
